```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
# Configure the map plot
import hvplot.pandas
import holoviews as hv
from holoviews import opts
#Scikit-learn modules
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
import seaborn as sns

```

```python
# Load the CSV file into a Pandas DataFrame
cars_sold_df = pd.read_csv("Outputs/cars_sold_uk.csv")

# Display sample data
cars_sold_df.head()
```

```python
# counting car makes
car_make_counts = cars_sold_df['make'].value_counts()

# Display the counts
print(car_make_counts)
```

```python
car_make_counts_df = car_make_counts.reset_index()
car_make_counts_df.columns = ['make', 'Count']

# Display the DataFrame
car_make_counts_df.head(9)
```

```python
# Calculate the correlation matrix
corrMatrix = cars_sold_df.corr()

# Plot the correlation matrix using a heatmap
sns.heatmap(corrMatrix, annot=True, cmap="viridis")

# Show the plot
plt.show()
```

```python
#Histogram
#This will provide a visual representation of the distribution of numeric features in our dataset
cars_sold_df.hist(figsize=(12, 10), bins=20)
plt.suptitle("Histograms of Numeric Features", y=1.02)
plt.show()
```

```python
# Boxplots
# This can helps identify outliers and understand the spread of car prices for different makes
sns.boxplot(x='make', y='price', data=cars_sold_df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Pair plots
#A pair plot displays scatterplots for numerical variables and histograms for each variable on the diagonal. 
# It's useful for identifying relationships between variables.
sns.pairplot(cars_sold_df, height=2)
plt.show()
```

```python
# Count plots
#This helps to understand the distribution of categorical variables, such as the count of cars for each fuel type.
sns.countplot(x='fuelType', data=cars_sold_df)
plt.show()
```

```python
# Violin plot
#Violin plots show the distribution of numeric data across different categories. 
#This example visualizes the distribution of prices based on transmission type.
sns.violinplot(x='transmission', y='price', data=cars_sold_df)
plt.show()
```

```python
#Scatter Plots
# Scatter plots can reveal relationships between two numeric variables, helping to understand how they are correlated.
plt.scatter(cars_sold_df['mileage'], cars_sold_df['price'])
plt.title('Scatter Plot of mileage vs. price')
plt.xlabel('mileage')
plt.ylabel('price')
plt.show()
```

```python
#Bar Plots
#Bar plots useful for visualizing the average price of cars for different years.
sns.barplot(x='year', y='price', data=cars_sold_df)
plt.xticks(rotation=90)
plt.show()
```

```python
#Distribution plot
#This plot provide insights into the distribution of the target variable, in this case, the prices of used cars.
sns.distplot(cars_sold_df['price'])
plt.title('Distribution of Car Prices')
plt.show()
```
